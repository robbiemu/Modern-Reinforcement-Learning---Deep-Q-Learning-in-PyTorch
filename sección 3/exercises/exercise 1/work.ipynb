```python
import gym
from trainer import Trainer
from agent import Agent
import matplotlib.pyplot as plt
```

```python
total_episodes = 1000

env = gym.make("FrozenLake-v0")
env.reset()                    
env.render()
print("\nAction space: ", env.action_space)
print("Observation space: ", env.observation_space)
```

```python
n_episodes = 10000
gamma = 0.99
def lambda_epsilon(x): return x - 1e-5 if x - 1e-5 > 0.01 else 0.01

# env, n_episodes, gamma, lambda_epsilon, alpha=0.001, epsilon=1
agent = Agent(env, n_episodes, gamma, lambda_epsilon)
trainer = Trainer(agent, env, n_episodes)
rewards = trainer.train()

plt.plot(rewards)
plt.show()
```
